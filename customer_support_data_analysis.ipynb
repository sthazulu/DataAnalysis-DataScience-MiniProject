```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("customer_support_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
issue_counts = df["Issue_Category"].value_counts()
issue_counts
```

```python
issue_counts.plot(kind="bar")
plt.title("Most Common Issue Categories")
plt.xlabel("Issue Category")
plt.ylabel("Count")
plt.show()
```

```python
avg_resolution = df.groupby("Issue_Category")["Resolution_Time_Hours"].mean()
avg_resolution
```

```python
avg_resolution.plot(kind="bar")
plt.title("Average Resolution Time by Issue Category")
plt.xlabel("Issue Category")
plt.ylabel("Avg Resolution Time (Hours)")
plt.show()
```

```python
print("Top Insights:")

print(f"- Most common issue: {issue_counts.idxmax()}")
print(f"- Issue with longest average resolution time: {avg_resolution.idxmax()} ({avg_resolution.max()} hrs)")
print(f"- Average satisfaction rating: {df['Satisfaction_Rating'].mean():.2f}")
```
